# **Exploratory Data Analysis**

### **Gender**

```python
ax = sns.countplot(x = 'gender',data = df,hue = 'gender' , palette = 'bwr_r',legend = False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['gender'], as_index = False)['price'].sum().sort_values(by = 'price', ascending = False)

sns.barplot(x = 'gender', y = 'price', data = sales_gen,hue = 'gender' , palette = 'bwr_r',legend = False)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### **Age**

```python
ax = sns.countplot(data = df, x = 'age group', hue = 'gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['age group'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)

sns.barplot(x = 'age group',y= 'price' ,data = sales_age,hue = 'age group' , palette = 'rainbow',legend = False)
```

From above graphs we can see that most of the buyers are of age group 55+ yrs female

### **Category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'category', hue = 'category' , palette = 'rainbow',legend = False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat = df.groupby(['category'], as_index=False)['price'].sum().sort_values(by='price', ascending=False).head(10)

sns.set(rc={'figure.figsize':(12,7)})
sns.barplot(data = sales_cat, x = 'category',y= 'price',hue = 'category' , palette = 'rainbow',legend = False)
```

From above graphs we can see that most of the sold products are from Clothing, food and cosmetics category and the most amount is spent on clothing, shooes and technology category.


### **Payment Method**

```python
sns.set(rc={'figure.figsize':(12,5)})
ax = sns.countplot(data = df, x = 'payment_method', hue = 'payment_method' , palette = 'rainbow',legend = False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['payment_method'], as_index=False)['price'].sum().sort_values(by='price', ascending=False)

sns.barplot(x = 'payment_method',y= 'price' ,data = sales_age,hue = 'payment_method' , palette = 'rainbow',legend = False)
```
